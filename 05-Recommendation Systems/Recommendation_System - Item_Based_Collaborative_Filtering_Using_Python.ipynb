```python
# Item based collaborative filtering python

#importing neccesary packages

import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Step 1 - Data Preparation

#Reading movielens data from 
#https://grouplens.org/datasets/movielens/

my_df = pd.read_csv('E:/datafiles/ml-100k/u.data', sep='\t', names=['user_id','item_id','rating','timestamp'])
```

```python
#Checking Data

my_df.head()
```

```python
#Importing movie title dataset and joining with main data my_df dataset

#user_id - the ID of the user who rated the movie.
#item_id - the ID of the movie.
#rating - The rating the user gave the movie, between 1 and 5.
#timestamp - The time the movie was rated.
#title - The title of the movie.

movie_titles = pd.read_csv('E:/datafiles/Movie_Titles.csv',encoding= 'unicode_escape')
my_df = pd.merge(my_df, movie_titles, on='item_id')
my_df.head()
```

```python
#Step 2 - Data exploration

#Statitical features of the movie

my_df.describe()
```

```python
#creating mean ratings data

ratings = pd.DataFrame(my_df.groupby('title')['rating'].mean())
ratings.head()
```

```python
#creating number of ratings data

ratings['number_of_ratings'] = my_df.groupby('title')['rating'].count()
ratings.head()
```

```python
#Plotting the jointplot

#Here the ratings ranges from 1, 2, 3, 4, 5
#x-axis ----> ratings
#y-axis ----> number_of_ratings
#1.Histogram1 gives the ratings of the movies from 1 to 5
#2.Histogram2 gives the density b/w rating vs number_of_ratings

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.jointplot(x='rating', y='number_of_ratings', data=ratings)
```

```python
#Creating User-Item interaction matrix

#We are creating this User-Item interaction matrix by using the pivot_table on my dataset my_df

movie_matrix_UII = my_df.pivot_table(index='user_id', columns='title', values='rating')
movie_matrix_UII.head()
```

```python
#Most rated movies,Now we are taking most rated movies for our recommended

ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
#Making recommendation - Example movie- FARGO

#Fetching ratings for Fargo
#This show different users ratings on FARGO movie

Fargo_user_rating = movie_matrix_UII['Fargo (1996)']
```

```python
Fargo_user_rating.head(20)
```

```python
#Finding the correlation FARGO with different movies

similar_to_fargo = movie_matrix_UII.corrwith(Fargo_user_rating)
```

```python
#Observing the result of similar movies correlated with FARGO 
#The correlation value ranges between -1 to +1

similar_to_fargo.head()
```

```python
#What we are interested is we want to make most meaning recommendations we will be doing this by

# creating a threshold for minimum number of ratings

#creating dataframe to bring in Number of ratings

corr_fargo = pd.DataFrame(similar_to_fargo, columns=['Correlation'])
corr_fargo.dropna(inplace = True)
corr_fargo.head()
```

```python
#Bringining in number_of_ratings into the corr_fargo dataframe

corr_fargo = corr_fargo.join(ratings['number_of_ratings'])
corr_fargo.head()
```

```python
#Now we are putting a filter on number_of_ratings column where the condition we are using is:
#Give me the movies where number_of ratings > 30

corr_fargo[corr_fargo['number_of_ratings'] > 30].sort_values(by='Correlation', ascending=False).head(10)
```
