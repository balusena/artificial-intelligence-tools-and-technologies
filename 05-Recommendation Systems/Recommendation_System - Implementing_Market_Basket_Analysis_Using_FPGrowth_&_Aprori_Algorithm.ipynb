```python
#Apriori Algorithm:

#Implementing Apriori Algorithm in python

#Create Database of transactions:----------------------> use this first

dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple','Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
#The apriori algorithm expects data in a One_Hot_Encoded pandas dataframe

#mlxtend(Machine_Learning_Extensions)  is a python library of useful tools of day-day data science tasks

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
#As it is difficult to understand the above One_Hot_Encoding notation we will convert this into dataframe

df = pd.DataFrame(te_ary)
df
```

```python
#By default, we get the column indicies of the items.For better readability,we can set use_colnames = True
#to convert these integer values into the respective item names.

df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support = 0.6, use_colnames = True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.8)
rules
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
#Viewing Top 10 rules

#The total rules depends upon the volume of transactions we are using for recommendation_rule_system

rules.head(10)
```
