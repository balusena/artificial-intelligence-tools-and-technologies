# Data Visualisation with Seaborn

```python
# Seaborn library

# Used for data visualization & based on Matplotlib
# Seaborn allows the creation of statistical graphics

# Functionalities

# Allows comparision between multiple variables
# Supports multi-plot grids
# Available univariate and bivariate visualizations
# Availability of different colour palettes
# Estimates and plots linear regression automatically
```

```python
# Seaborn vs Matplotlib

# Seaborn fixes two short commings of Matplotlib

# 1.Matplotlib settings are difficult to figure out.Seaborn comes with numerous customized themes and high level interfaces
# 2.Matplotlib doesnot serve well when it comes to dealing with DataFrames,while Seaborn functions actually work on DataFrames
```

```python
# Python Seaborn Functions

# Visualizing Statistical Relationships
# Process of understanding relationships between variables of a dataset
b
# Plotting with Categorical Data
# Main variable is further divided into discrete groups

# Visualizing the distribution of a dataset
# Understanding datasets with context to being univariate or bivariate
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# https://github.com/mwaskom/seaborn-data

a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", data=a)
```

```python
# scatter-plot

a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", hue="year", data=a)
```

```python
# line-plot

b = sns.load_dataset("tips")
sns.relplot(x="time", y="tip", data=b, kind="line")
```

```python
# Plotting with categorical data

sns.catplot(x="day", y="total_bill", data=b)
```

```python
sns.catplot(x="day", y="total_bill", data=b, kind="violin")
```

```python
sns.catplot(x="day", y="total_bill", data=b, kind="boxen")
```

```python
# Visualizing the distribution of a dataset

# Plotting univariate and bivariate distributions

# Example for univariate distribution

from scipy import stats

c = np.random.normal(loc=5, size=100, scale=2)
sns.distplot(c)
```

```python
# Example for bivariate distribution

x=pd.DataFrame({"Day":[1,2,3,4,5,6,7],"Grocery":[30,80,45,23,51,46,76],"Clothes":[13,40,34,23,54,67,98],"Utensils":[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7])
y=pd.DataFrame({"Day":[8,9,10,11,12,13,14],"Grocery":[30,80,45,23,51,46,76],"Clothes":[13,40,34,23,54,67,98],"Utensils":[12,32,27,56,87,54,34]},index=[8,9,10,11,12,13,14])
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="kde", color="b");               
```

```python
# Multi-plot Grids (FacetGrid_function,pairgrid_function)

# Graphs are plotted side by side using the same scale and axes to aid comparision

# Example for FacetGrid_function

a=sns.load_dataset("iris")
b=sns.FacetGrid(a, col="species")
b.map(plt.hist,"sepal_length")
```

```python
#when we have pair of variables to compare use PairGrid_function

# Exaple for PairGrid_function

a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
# Plot Aesthetics

# Deals with making our plots more attractive and delightful: using various themes and color-pallettes

sns.set(style="darkgrid")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="dark")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
# We can also make changes to our plots such as Adding or removing axes,colors etc...

# How to remove axes lines that are present in our graph 

sns.set(style="white", color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=a)
```

```python
# How to remove axes lines that are present in our graph 

sns.set(style="white", color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=a)
sns.despine(offset=10, trim=True)
```

```python
# Seaborn provides number of color pallettes if you want to check them use color palletes function
# for deeper variations we can use hls_pallete,husl_pallete etc

c=sns.color_palette()
sns.palplot(c)
```

```python
# strip plot

b = sns.load_dataset("tips")
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=b, kind="strip", height=4, aspect=.7);
```

```python
# swarm plot

b = sns.load_dataset("tips")
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=b, kind="swarm", height=4, aspect=.7);
```

```python
# Box plot

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
b = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="box", data=b);
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
b = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="violin", data=b);
```

```python
# Boxen plot

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
b = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="boxen", data=b); 
```

```python
# Point plot

b = sns.load_dataset("tips")
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=b, kind="point", dodge=True, height=4, aspect=.7);
```

```python
# Bar plot

b = sns.load_dataset("tips")
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=b, kind="bar", height=4, aspect=.7);
```

```python
# Count plot

b = sns.load_dataset("titanic")
sns.catplot(x="class", hue="who", col="survived", data=b, kind="count", height=4, aspect=.7);
```

```python
# set the background style of the plot
# Distplot

b = sns.load_dataset("tips") 
sns.set_style('whitegrid') 
sns.distplot(b['total_bill'], kde = False, color ='red', bins = 30)
```

```python
# Joinplot

b = sns.load_dataset("tips")
sns.jointplot(x ='total_bill', y ='tip', data = b)
```

```python
# KDE shows the density where the points match up the most 

sns.jointplot(x ='total_bill', y ='tip', data = b, kind ='kde') 
```

```python
# Joinplot

b = sns.load_dataset("tips")
sns.pairplot(b, hue ="sex", palette ='coolwarm')
```

```python
# Rugplot

b = sns.load_dataset("tips")
sns.rugplot(b['total_bill'])
```
