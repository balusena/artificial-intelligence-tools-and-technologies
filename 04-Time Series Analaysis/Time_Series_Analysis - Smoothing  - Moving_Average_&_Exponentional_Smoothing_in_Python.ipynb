```python
#Smoothing Techniques Discussed here:-
#1.Moving Average
#2.Simple Smoothing
#3.Holt Method/Triple smoothing

#Importing Packages and data

import pandas as pd
from matplotlib import pyplot
from matplotlib import pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from pylab import rcParams
rcParams['figure.figsize'] = 20, 5
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
electrictytimeseries = pd.read_csv("E:\datafiles\Electric_Production.csv", header = 0, index_col = 0)
```

```python
electrictytimeseries.shape
```

```python
electrictytimeseries.head()
```

```python
#Moving Average Method

plt.plot(electrictytimeseries[1:50]["Units"])
plt.xticks(rotation = 30)
plt.show()
```

```python
#Rolling Average Transform

#By using Rolling Average we can find Mean and Varience of Time Series Data

rollingseries = electrictytimeseries[1:50].rolling(window = 5)
rollingmean = rollingseries.mean() #finding rolling mean MA(5)

print(rollingmean.head(10))
```

```python
#plot transformed dataset

rollingmean.plot(color = 'red')
pyplot.show()
```

```python
rollingseries = electrictytimeseries[1:50].rolling(window = 10)
rollingmean = rollingseries.mean() #finding rolling mean MA(5)

print(rollingmean.head(10))
```

```python
#plot transformed dataset

rollingmean.plot(color = 'red')
pyplot.show()
```

```python
#1.Simple Exponential Smoothing or Single Exponential Smoothing

data = electrictytimeseries[1:50]

fit1 = SimpleExpSmoothing(data).fit(smoothing_level = 0.2, optimized = False)

fit2 = SimpleExpSmoothing(data).fit(smoothing_level = 0.8, optimized = False)
```

```python
plt.figure(figsize = (18,8))
plt.plot(electrictytimeseries[1:50], marker = 'o', color = 'black')
plt.xticks(rotation = 30)
plt.plot(fit1.fittedvalues, marker = 'o', color = 'blue')
plt.plot(fit2.fittedvalues, marker = 'o', color = 'red')
```

```python
#Holt Exponential Smoothning Method or Triple Exponential Smoothing Method

fit1 = Holt(data).fit() #linear trend

fit2 = Holt(data, exponential = True).fit() #exponential trend
```

```python
plt.plot(data, marker = 'o', color = 'black')
plt.xticks(rotation = 30)
plt.plot(fit1.fittedvalues, marker = 'o', color = 'blue')
plt.plot(fit2.fittedvalues, marker = 'o', color = 'red')
```
