```python
#TF-IDF stands for Term frequency-Inverse Document Frequency. The TF-IDF weight is a weight often used in information 
#retrieval and text mining. Variations of the TF-IDF weighting scheme are often used by search engines in scoring and 
#ranking a document's relevance given a query.The terms with higher weight scores are considered to be more important.

#Typically, the TF-IDF weight is composed by two terms-

#1.Normalized Term Frequency/Term Frequency (TF)
#2.Inverse Document Frequency (IDF)

#Formula for Term Frequency (TF) = [No of repetition of words in sentence/No of words in sentence]

#Formula for Inverse Document Frequency (IDF) = log(No of Total sentences/No of sentences contaning the actual words)

#Formula for TF-IDF Scoring = TF*IDF

#Note:-This TF-IDF works very well with huge amount of datasets bcz in this approach weightage of words plays major role.


#Import the nltk

import nltk
```

```python
paragraph =  """I have three visions for India. In 3000 years of our history, people from all over 
               the world have come and invaded us, captured our lands, conquered our minds. 
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours. 
               Yet we have not done this to any other nation. We have not conquered anyone. 
               We have not grabbed their land, their culture, 
               their history and tried to enforce our way of life on them. 
               Why? Because we respect the freedom of others.That is why my 
               first vision is that of freedom. I believe that India got its first vision of 
               this in 1857, when we started the War of Independence. It is this freedom that
               we must protect and nurture and build on. If we are not free, no one will respect us.
               My second vision for India’s development. For fifty years we have been a developing nation.
               It is time we see ourselves as a developed nation. We are among the top 5 nations of the world
               in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.
               Our achievements are being globally recognised today. Yet we lack the self-confidence to
               see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect?
               I have a third vision. India must stand up to the world. Because I believe that unless India 
               stands up to the world, no one will respect us. Only strength respects strength. We must be 
               strong not only as a military power but also as an economic power. Both must go hand-in-hand. 
               My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of 
               space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.
               I was lucky to have worked with all three of them closely and consider this the great opportunity of my life. 
               I see four milestones in my career"""
```

```python
# Cleaning the texts
#Downloading all the stopwords content from nltk.corpus and verfying with the stopwords in paragraph
#In the below paragrah we will see words such as of,them,the,our,of,then etc., these words doesnt do anything when we are
#performing Positive & Negative Sentiment Analysis hence we use stopwords on this paragraph to remove these words

#Tokenizing sentences from the above paragrah/corpora.
#We are using PorterStemmer for stemming purpose
#We are using WordNetLemmatizer for lemmatization purpose

import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
#We are initializing PorterStemmer() and creating an object ps for it.
#We are initializing WordNetLemmatizer() creating an object wordnet for it.
#We are converting the paragraph into sentences by using nltk.sent_tokenize(paragraph).
#Now create a list i.e, corpus[] so in this list we are storing all the sentences that are cleaned by using below processes.
#This shows the difference between sentences and corpus[]

# Dataset Preprocessing

#Now sending all the sentences through a loop using the function corpus
#First removing all the special characters other than [^a-zA-Z] by using regular expression
#Then lowering the words of each & every sentences,then spliting and passing all list of words through the list compherension.
#Through this we are doing two things one is removing stopwords present in english and second is stemming of that word
#After performing all these nltk operations we are again joining these words to review back to list corpus[] 


ps = PorterStemmer()
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#Instead of using stemming for the paragraph we will use lemmatization and see the diffrence in the word conversion.

ps = PorterStemmer()
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#These are all my words after applying nltk(for removing lemmatization,stopwords) from the messages

corpus
```

```python
# Creating the TF-IDF model
# Creating TF-IDF Document Matrix and finally converting into an array

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
X = cv.fit_transform(corpus).toarray()
df= pd.DataFrame(X, columns=cv.get_feature_names())
```
