```python
#2.Feature Selection- With Correlation

#In this step we will be removing the features which are highly correlated

#importing libraries

from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the dataset

data = load_boston()
df = pd.DataFrame(data.data, columns = data.feature_names)
df["MEDV"] = data.target
```

```python
df.head()
```

```python
data.feature_names
```

```python
#If an independent feature is highly correlated with dependent feature there is no need to remove those
#features these features will play a very important role while training our machine learning model
#but if we have three independent faetures which are highly correlated with each other by morethan 90%
#then form those three independent features we will only use one independent feature to reudce the
#correlation by dropping another two independent features

#Here X ---> indepenedent features 'CRIM','ZN','INDUS' Y ----> depenedent fetaure 'MEDV'


X = df.drop("MEDV",axis=1) 
y = df["MEDV"]
```

```python
df.head()
```

```python
X.head()
```

```python
#separate the dataset into train,test and split to overcome the overfitting problem
#all the correlation functionalities will be done on the X_train dataset 
#whatever changes we will be doing on X_train dataset will be done on X_test dataset
#suppose we find 4 features that are highly correlated,we will directly remove from X_train
#then in X_test we will not do correlatin retest again on X_test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

```python
#The correlation allways ranges from -1 to +1

X_train.corr()
```

```python
import seaborn as sns
#Using Pearson Correlation

plt.figure(figsize=(12,10))
cor = X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
#Here we are applying the corr_feature to the X_train with some threshold-0.7 value but not to the X_test 

corr_features = correlation(X_train, 0.7)
len(set(corr_features))
```

```python
corr_features
```

```python
#Here we are applying the corr_feature to the X_train but not to the X_test 
#we are directly dropping the corr_features from X_test.

X_train.drop(corr_features,axis=1)
X_test.drop(corr_features,axis=1)
```
