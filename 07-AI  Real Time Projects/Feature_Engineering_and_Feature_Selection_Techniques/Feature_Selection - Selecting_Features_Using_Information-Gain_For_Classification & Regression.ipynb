```python
#3.Mutual Information:-

#Mutual Information is nothing but information gain we specifically use in decission trees.

#MI Estimate mutual information for a discrete target variable.

#Mutual information (MI) between two random variables is a non-negative value, which measures the 
#dependency between the variables. It is equal to zero if and only if two random variables are 
#independent, and higher values mean higher dependency.

#The function relies on nonparametric methods based on entropy estimation from k-nearest neighbors
#distances.

#Inshort

#A quantity called mutual information measures the amount of information one can obtain from one 
#random variable given another.

#The mutual information between two random variables X and Y can be stated formally as follows:

#I(X ; Y) = H(X) – H(X | Y) Where I(X ; Y) is the mutual information for X and Y, H(X) is the entropy 
#for X and H(X | Y) is the conditional entropy for X given Y. The result has the units of bits.

#Youtube Videos:-

#1.Entropy: https://www.youtube.com/watch?v=1IQOtJ4NI_0

#2.Information Gain: https://www.youtube.com/watch?v=FuTRucXB9rA

#3.Gini Impurity: https://www.youtube.com/watch?v=5aIFgrrTqOw

#4.Statistical test: https://www.youtube.com/watch?v=4-rxTA_5_xA
#                    https://www.youtube.com/watch?v=YrhlQB3mQFI
```

```python
#Feature Selection-Information gain - mutual information In classification Problem Statements

import pandas as pd

df=pd.read_csv('https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv')
df.head()
```

```python
#To get the different types of wines in wines dataset

df['Wine'].unique()
```

```python
df.info()
```

```python
### Train test split to avoid overfitting

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['Wine'], axis=1),
    df['Wine'],
    test_size=0.3,
    random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.feature_selection import mutual_info_classif

# determine the mutual information
# This is the library mutual_info_classif which is present inside sklearn used for classification problem
# Care should be taken to handle null values before applying this mutual_info_classif
# After applying this  mutual_info_classif we will never get negative value, value will be 0 or +ve-no's
# If you are getting a value very higher we can say that the dependency of this variable on target is more
# From below array which ever value is higher is the most important feature, here high value is 0.71254615

mutual_info = mutual_info_classif(X_train, y_train)
mutual_info
```

```python
# There are so many features,we are converting them into series by using pd.series
# We are taking all the columns in X_train and assigning them to index
# Now sorting this values in ascending order i.e, from highest- 'Flavanoids 0.712546' to lowest

mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
#let's plot the ordered mutual_info values per feature

mutual_info.sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

```python
from sklearn.feature_selection import SelectKBest

#Now we Will select the  top 5 important features

sel_five_cols = SelectKBest(mutual_info_classif, k = 5)
sel_five_cols.fit(X_train, y_train)
X_train.columns[sel_five_cols.get_support()]
```

```python
#Difference Between Information Gain And Mutual Information

#MI(X ; Y) = H(X) – H(X | Y) and IG(S, a) = H(S) – H(S | a)

#As such, mutual information is sometimes used as a synonym for information gain. Technically, they 
#calculate the same quantity if applied to the same data.
```

```python
#2.Feature Selection-Information gain - mutual information In Regression Problem Statements.

#It will calculate the relationship betweeen our independent feature X and dependent feature y

#Where our dependent feature is a continuouus target variable

#https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

import pandas as pd
housing_df = pd.read_csv('E:/datafiles/loan.csv')
```

```python
#We are selecting all the continuous features i.e integer,float datatypes to avoid encoding techniques
#For this selected numerical features we will find mutual information gain along with output feature

numeric_lst=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_cols = list(housing_df.select_dtypes(include=numeric_lst).columns)
```

```python
numerical_cols
```

```python
#For all these numerical features we will apply the mutual information gain


housing_df = housing_df[numerical_cols]
```

```python
housing_df.head()
```

```python
#We are dropping the id column which not in use in dataset

housing_df = housing_df.drop("Id",axis=1)
```

```python
### It is always a good practice to split train and test data to avoid overfitting

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(housing_df.drop(labels=['SalePrice'], axis=1),
    housing_df['SalePrice'],
    test_size=0.3,
    random_state=0)
```

```python
X_train
```

```python
#To check whether X_train is having any null values are not

X_train.isnull().sum()
```

```python
#We are applying mutual information gain on regression by using mutual_info_regression.
#To avoid data preprocessing we are making X_train.fillna(0)
#All the values in array will be greater than zero and less than 1(>0 & <1)
#The higher the value in array,then it is more dependent, to lower value to less depenedent to output

from sklearn.feature_selection import mutual_info_regression

# determine the mutual information

mutual_info = mutual_info_regression(X_train.fillna(0), y_train)
mutual_info
```

```python
#Now wew arranging these above array values in series and making in ascending order with index

mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
#Now plotting the mutual_info.sort 

mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
#Now selecting top percentile features present in mutual_info variable by using SelectPercentile

from sklearn.feature_selection import SelectPercentile
```

```python
#Selecting the top 20 percentile

selected_top_columns = SelectPercentile(mutual_info_regression, percentile=20)
selected_top_columns.fit(X_train.fillna(0), y_train)
```

```python
selected_top_columns.get_support()
```

```python
#To select top 20% of feature which are having high mutual information gain
#From below we can conclude that upto 'GarageArea' 20% features are having high information gain when 
#compared to other 80% of features are having less information gain less than compared to 'GarageArea'

X_train.columns[selected_top_columns.get_support()]
```
