```python
# what is cross validation and what are its different types

#Cross-validation is primarily used in applied machine learning to estimate the skill of a machine learning
#model on unseen data.That is,to use a limited sample in order to estimate how the model is expected to 
#perform in general when used to make predictions on data not used during the training of the model.

#1. Leave  P out cross validation(in this the P value is selected from 1,2,.......,n)
#2. K fold Cross Validation
#3. Stratified K fold Cross Validation
#4. Time Series Cross Validation

#Cross Validations Techniques

import pandas as pd
df=pd.read_csv('E:/datafiles/cancer_dataset.csv')
df.head()
```

```python
# Independent And dependent features

X=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
X.head()
```

```python
X=X.dropna(axis=1)
```

```python
X.head()
```

```python
y.value_counts()
```

```python
# 1.HoldOut Validation Approach- Train And Test Split

#in this technique we usually do train_test_split on a dataset nothing but HoldOut Validation technique
#we train our model with train dataset and we validate our model with test dataset.

#We are doing this train_test_split by using random_sate parameter we are not sure how this data is taken
#from dataset and this may lead to overfiiting or underfitting problem by using this technique.
#when we change the random_state parameter the accuracy will change all the time in order to overcome this
#problem we are using the k-fold cross validation technique.

#This tecnnique mainly used for balanced datasets.

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=4)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
result = model.score(X_test, y_test)
print(result)
```

```python
# 2.K Fold Cross Validation

#In this technique we will select a K value based on the k value the tain & test data will be decided
#similarly this process will be repeated for the all the k values and even we can take the mean of all
#outcomes and show it to our stake holders both the min and max values generated by this technique.

#This tecnnique mainly used for balanced datasets.

from sklearn.model_selection import KFold
model=DecisionTreeClassifier()
kfold_validation=KFold(10)

import numpy as np
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,y,cv=kfold_validation)
print(results)
print(np.mean(results))
```

```python
# 3.Stratified K-fold Cross Validation

#In this technique we will select a K value based on the k value the train & test data will be decided
#similarly this process will be repeated for the all the k values,in this technique it will maintain 
#good proportion of data from both train and test data to overcome the problem that occurs in k-fold 
#cross validation hence this produces good accuracy when compared to k-fold cross validation.

#This tecnnique mainly used for imbalanced dataset.

from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
scores=cross_val_score(model,X,y,cv=skfold)
print(np.mean(scores))
scores
```

```python
# 4.Leave One Out Cross Validation(LOOCV)

#In this validation technique only one test data will be selected and remaining will be taken as train data
#similarliy this will be repeated for all records in the dataset in the same way hence it increases the
#n number of iterations and leads to computation overhead this is the main drawback and even it lead to 
#low bias this leads to low in accuracy and increase in high error for new train and test data.

#This tecnnique mainly used for balanced datasets.

from sklearn.model_selection import LeaveOneOut
model=DecisionTreeClassifier()
leave_validation=LeaveOneOut()
results=cross_val_score(model,X,y,cv=leave_validation)
results
```

```python
print(np.mean(results))
```

```python
# 5.Repeated Random Test-Train Splits

#This technique is a hybrid of traditional train-test splitting and the k-fold cross-validation method. 
#In this technique,we create random splits of the data in the training-test set manner and then repeat 
#the process of splitting and evaluating the algorithm multiple times, just like the cross-validation 
#method

from sklearn.model_selection import ShuffleSplit
model=DecisionTreeClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
results=cross_val_score(model,X,y,cv=ssplit)
results
```
