```python
#Word Emdedding Techniques Using Embedding Layer in Keras
#Libraries used Tensorflow <2.0 and Keras
#Libraries used Tensorflow > 2.0 and Keras

#from keras.preprocessing.text import one_hot  Tensorflow<2.0

from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences

sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
### Vocabulary size from the dictionary of words created

voc_size=10000
```

```python
#One Hot Representation

#Which is used to create an index from the dictionary of words for each unique word

onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

```python
#Word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
#The index obtained from one hot encoding of words to embedded layer keras to form embedding matrix
#We are taking sent_length=8, by default and adding additional zeros with padding(pre(starting)) for equal sentance flow

sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
#dimension of the features representation

dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
